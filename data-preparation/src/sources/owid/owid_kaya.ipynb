```python
import pandas as pd

# Read CSV file into a DataFrame
df = pd.read_csv('files/owid_kaya-identity-co2.csv')
df
```

```python
import matplotlib.pyplot as plt

df.query("Code=='FRA'").set_index("Year")[["Annual CO₂ emissions per unit energy (kg per kilowatt-hour)", "Annual CO₂ emissions per GDP (kg per international-$)"]].plot(kind = "line",figsize = (10,3))
plt.show()
```

```python
# Filter the data for France and extract the relevant columns
france_data = df.query("Code=='FRA'").set_index("Year")[["GDP per capita"]]

# Create a line plot of the data
ax = france_data.plot(kind="line", figsize=(10, 5))

# Set the tick frequency and format for the y-axis
ax.yaxis.set_major_locator(plt.MaxNLocator(12))
ax.yaxis.set_major_formatter('${x:,.2f}')

# Set the axis labels and title
ax.set_xlabel('Year')
ax.set_ylabel('GDP per capita')
ax.set_title('GDP per capita in France')

plt.show()
```
