```python
import requests
import re
def _pattern():
    search = 'bp-stats-review*dataset-narrow-format.csv'
    search = search.replace('.','@@').replace('*','.*').replace('@@','\.')
    print(search)
    return r'.*' + search 

# URL of the webpage
url = 'https://www.bp.com/en/global/corporate/energy-economics/statistical-review-of-world-energy/downloads.html'

# Make a GET request to retrieve the HTML content of the page
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract all links from the HTML content that match a specific pattern
    pattern = r'href=["\'](.*?)["\']'
    links = re.findall(pattern, response.text)

    # Filter the links based on the pattern
    filtered_links = [link for link in links if re.match(_pattern(), link)]

    # Print the filtered links
    for link in filtered_links:
        print(link)
else:
    print(f'Request failed with status code: {response.status_code}')
```
